```python
def visualize_tokenization_improved(text, tokenizer):
    # Get the original text and lowercase it for better matching with BERT-uncased
    original_text = text
    original_lowercase = text.lower()

    # Tokenize without special tokens
    tokens = tokenizer.tokenize(text)

    # Create a DataFrame for visualization
    token_data = []
    current_pos = 0
    last_token_end = 0

    for i, token in enumerate(tokens):
        # Remove ## from the beginning of tokens if present
        clean_token = token.replace('##', '')

        # Check if this is a subword continuation
        if token.startswith('##'):
            # For subword tokens, they continue from the previous position
            token_start = last_token_end
        else:
            # For new words, search in the remaining text
            # Use a more robust search method that accounts for lowercase and skips already processed text
            search_start = current_pos
            token_start = original_lowercase.find(clean_token, search_start)

            # If not found, try alternative approaches
            if token_start == -1:
                # Check if it's a special character or punctuation
                if len(clean_token) == 1 and not clean_token.isalnum():
                    token_start = original_lowercase.find(clean_token, search_start)

                # If still not found, try to find closest match
                if token_start == -1:
                    # Just use the current position as a fallback
                    token_start = current_pos

            # Update the current position for next searches
            current_pos = token_start + len(clean_token)

        token_end = token_start + len(clean_token)
        last_token_end = token_end

        # Extract the original text segment (maintaining case)
        if 0 <= token_start < len(original_text) and 0 <= token_end <= len(original_text):
            original_segment = original_text[token_start:token_end]
        else:
            original_segment = "?"  # Out of bounds

        token_data.append({
            'Token': token,
            'ID': tokenizer.convert_tokens_to_ids([token])[0],
            'Start': token_start,
            'End': token_end,
            'Original Text': original_segment
        })

    # Add post-processing to fix overlaps or gaps
    for i in range(1, len(token_data)):
        if token_data[i]['Start'] < token_data[i-1]['End']:
            # Fix overlaps
            token_data[i]['Start'] = token_data[i-1]['End']
            token_data[i]['End'] = token_data[i]['Start'] + len(token_data[i]['Token'].replace('##', ''))
            token_data[i]['Original Text'] = original_text[token_data[i]['Start']:token_data[i]['End']]

    return pd.DataFrame(token_data)

# Example usage
sample_text = "Le modèle BERT utilise la tokénisation WordPiece."
token_df = visualize_tokenization_improved(sample_text, tokenizer)
display(token_df)
```

Voici ce que contient généralement outputs pour un modèle BERT de base :

**outputs.last_hidden_state** : Un tenseur de forme (batch_size, sequence_length, hidden_size) qui représente les états cachés de la dernière couche du modèle pour chaque token.

**outputs.pooler_output** : Un tenseur de forme (batch_size, hidden_size) qui est le résultat du passage de l'état caché du token [CLS] (le premier token) à travers une couche linéaire suivie d'une activation tanh. C'est souvent utilisé comme représentation de la phrase entière.

```python
## 2. Encoding Sequences with BERT

def encode_and_visualize(texts, tokenizer, max_length=None):
    # Encode the texts
    encoding = tokenizer(
        texts,
        return_tensors='pt',
        padding=True,
        truncation=True if max_length else False,
        max_length=max_length
    )

    # Print the encoding components
    print(f"Input shape: {encoding['input_ids'].shape}")
    print(f"Attention mask shape: {encoding['attention_mask'].shape}")

    # Create DataFrames for visualization
    results = []

    for i, text in enumerate(texts):
        input_ids = encoding['input_ids'][i].tolist()
        attn_mask = encoding['attention_mask'][i].tolist()
        tokens = tokenizer.convert_ids_to_tokens(input_ids)

        # Create a DataFrame for this sequence
        df_data = []
        for j, (token_id, token, mask) in enumerate(zip(input_ids, tokens, attn_mask)):
            df_data.append({
                'Position': j,
                'Token': token,
                'Token ID': token_id,
                'Is Special': token in [tokenizer.cls_token, tokenizer.sep_token, tokenizer.pad_token],
                'Is Padding': token == tokenizer.pad_token,
                'Attention Mask': mask
            })

        df = pd.DataFrame(df_data)
        results.append(df)

        print(f"\nSequence {i+1}: \"{text}\"")
        display(df)

    # Visualize the attention mask
    fig, axes = plt.subplots(len(texts), 1, figsize=(10, 2 * len(texts)))
    if len(texts) == 1:
        axes = [axes]

    for i, ax in enumerate(axes):
        attn_mask = encoding['attention_mask'][i].numpy().reshape(1, -1)
        sns.heatmap(attn_mask, ax=ax, cmap='Blues', cbar=False,
                    xticklabels=tokenizer.convert_ids_to_tokens(encoding['input_ids'][i]),
                    yticklabels=['Attention'])
        ax.set_title(f"Attention Mask for Sequence {i+1}")

    plt.tight_layout()
    plt.show()

    return encoding, results

# Example with padding and different length sequences
example_texts = [
    "Un  texte court.",
    "Ceci est une séquence plus longue qui nécessitera davantage de tokens pour être correctement encodée."
]

encoding, dfs = encode_and_visualize(example_texts, tokenizer)
```

```python
## 4. Working with BERT Outputs and Hidden States

# Process text through BERT and examine outputs
def analyze_bert_outputs(text, tokenizer, model):
    # Tokenize input
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

    # Get BERT output with hidden states
    outputs = model(**inputs, output_hidden_states=True, output_attentions=True)

    # Examine output shapes
    print(f"Last hidden state shape: {outputs.last_hidden_state.shape}")
    print(f"Pooler output shape: {outputs.pooler_output.shape}")
    print(f"Number of hidden states: {len(outputs.hidden_states)}")
    print(f"Hidden state shapes: {outputs.hidden_states[0].shape}")
    print(f"Number of attention layers: {len(outputs.attentions)}")
    print(f"Attention shape for first layer: {outputs.attentions[0].shape}")

    # Return the output for further analysis
    return inputs, outputs

input_text = "L’hiver, nous irons dans un petit wagon rose"
inputs, outputs = analyze_bert_outputs(input_text, tokenizer, model)
```

```python
# Extract the hidden states for each token across all layers
def visualize_hidden_states(inputs, outputs, tokenizer):
    # Get the tokens
    input_ids = inputs['input_ids'][0].tolist()
    tokens = tokenizer.convert_ids_to_tokens(input_ids)

    # Get the hidden states
    hidden_states = outputs.hidden_states

    # Convert to numpy for easier manipulation
    # Shape: [layers, batch_size, seq_length, hidden_dim]
    hidden_states_np = np.array([hs.detach().numpy() for hs in hidden_states])

    # Reduce dimensionality for visualization (mean across hidden dimension)
    # This gives us a representation of the "activation" of each token in each layer
    token_activations = hidden_states_np.mean(axis=3)

    # Reshape to [layers, seq_length]
    token_activations = token_activations[:, 0, :]

    # Create a heatmap
    fig, ax = plt.subplots(figsize=(12, 8))
    sns.heatmap(token_activations,
                xticklabels=tokens,
                yticklabels=[f'Layer {i+1}' for i in range(len(hidden_states))],
                cmap='viridis',
                ax=ax)
    plt.title('Token Activations Across BERT Layers')
    plt.xlabel('Tokens')
    plt.ylabel('Layers')
    plt.tight_layout()
    plt.show()

    # Return processed data for further analysis
    return tokens, hidden_states_np

tokens, hidden_states_np = visualize_hidden_states(inputs, outputs, tokenizer)
```

```python
### 4.1 Visualizing Attention Patterns

def visualize_attention(inputs, outputs, tokenizer):
    # Get the attention weights
    # Shape: [layers, batch, heads, seq_length, seq_length]
    attention = outputs.attentions

    # Convert to numpy
    attention_np = np.array([attn.detach().numpy() for attn in attention])

    # Get tokens for better visualization
    input_ids = inputs['input_ids'][0].tolist()
    tokens = tokenizer.convert_ids_to_tokens(input_ids)

    # Select a layer and head to visualize
    layer_to_viz = 8  # 9th layer (0-indexed)
    head_to_viz = 4   # 5th attention head (0-indexed)

    # Create heatmap for the selected attention head
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        attention_np[layer_to_viz, 0, head_to_viz],
        xticklabels=tokens,
        yticklabels=tokens,
        cmap='Blues'
    )
    plt.title(f'Attention Pattern: Layer {layer_to_viz+1}, Head {head_to_viz+1}')
    plt.xlabel('Attended to')
    plt.ylabel('Attended from')
    plt.tight_layout()
    plt.show()

    return attention_np

attention_np = visualize_attention(inputs, outputs, tokenizer)
```

```python
## 5. Working with Multiple Sentences and Next Sentence Prediction

# For next sentence prediction functionality, we need to use a different model
from transformers import BertForNextSentencePrediction

nsp_model = BertForNextSentencePrediction.from_pretrained('bert-base-uncased')

def analyze_next_sentence_prediction(sentence_a, sentence_b, tokenizer, model):
    # Encode the sentence pair
    encoding = tokenizer(sentence_a, sentence_b, return_tensors='pt', padding=True)

    # Get NSP logits
    outputs = model(**encoding)
    logits = outputs.logits

    # Get prediction (0 = IsNext, 1 = NotNext)
    prediction = torch.softmax(logits, dim=1).detach().numpy()

    print(f"Sentence A: {sentence_a}")
    print(f"Sentence B: {sentence_b}")
    print(f"Probability that B follows A: {prediction[0][0]:.4f}")
    print(f"Probability that B does NOT follow A: {prediction[0][1]:.4f}")

    return prediction

# Example sentence pairs
sentence_pairs = [
    ("J'ai acheté un nouveau livre hier.", "J'ai commencé à le lire dès que je suis rentrée chez moi."),
    ("J'ai acheté un nouveau livre hier.", "Les oiseaux migrent vers le sud en hiver.")
]

for pair in sentence_pairs:
    analyze_next_sentence_prediction(pair[0], pair[1], tokenizer, nsp_model)
```
